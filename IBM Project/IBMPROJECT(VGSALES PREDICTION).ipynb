```python
data= pd.read_csv("C:\\Users\\I.SAI GOWTHAM REDDY\\Downloads\\vgsales.csv (1).zip")
sns.countplot(y='Genre',data=data,order=data['Genre'].value_counts().index)
plt.xticks(rotation=80); 
```

```python
numeric_columns = data.select_dtypes(include=['float64', 'int64'])

correlation_matrix = numeric_columns.corr()


plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".3f")
plt.title("Pairwise Correlation of the Video Game Sales Features")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Year', y='Genre', hue='Genre', palette='tab10')
plt.xlabel('Year')
plt.ylabel("Sales (in millions)")
plt.title("Video Games Sales Over the Years")
plt.legend(title='Sales Category', loc='upper left')
plt.grid(True)
plt.show()
```

```python
Data_Set['Year'] = pd.to_numeric(Data_Set['Year'], errors='coerce')

plt.figure(figsize=(12, 6))

bar_width = 0.2
index = Data_Set['Year'].unique()
bar_positions = [index - 2 * bar_width, index - bar_width, index, index + bar_width, index + 2 * bar_width]

plt.bar(bar_positions[0], Data_Set.groupby('Year')['NA_Sales'].sum(), bar_width, label='NA Sales')
plt.bar(bar_positions[1], Data_Set.groupby('Year')['EU_Sales'].sum(), bar_width, label='EU Sales')
plt.bar(bar_positions[2], Data_Set.groupby('Year')['JP_Sales'].sum(), bar_width, label='JP Sales')
plt.bar(bar_positions[3], Data_Set.groupby('Year')['Other_Sales'].sum(), bar_width, label='Other Sales')
plt.bar(bar_positions[4], Data_Set.groupby('Year')['Global_Sales'].sum(), bar_width, label='Global Sales')

plt.xlabel('Year')
plt.ylabel('Sales (in millions)')
plt.title('Video Game Sales Over the Years')
plt.xticks(index, rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
area_sales = Data_Set[["NA_Sales","EU_Sales", "JP_Sales", "Other_Sales"]]
area_sales = area_sales.melt(var_name='Area',value_name='Total_Sales')
area_sales = area_sales.groupby('Area')['Total_Sales'].sum().reset_index()

plt.figure(figsize=(15,10))
plt.pie(area_sales['Total_Sales'],labels=area_sales['Area'],colors=sns.color_palette('magma_r'),autopct='%.2f%%')
plt.show()
```
